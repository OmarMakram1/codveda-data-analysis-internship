```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cleaned_house_dataset.csv")
```

```python
df.head(5)
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(df.corr(),cmap='bwr',
            center=0, annot=True)
plt.title('Correlation Matrix')
plt.savefig("Correlation Matrix.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
r = df["RM"].corr(df["MEDV"])

fig, ax = plt.subplots(figsize=(10, 7))

sns.regplot(
    data=df, x="RM", y="MEDV",
    scatter_kws={"alpha": 0.4, "s": 40, "color": "#378ADD"},
    line_kws={"color": "#E24B4A", "linewidth": 2},
    ax=ax
)

ax.set_title("Rooms vs. Median Home Value ", fontsize=15, pad=15, fontweight="bold")
ax.set_xlabel("RM – Average number of rooms", fontsize=12)
ax.set_ylabel("MEDV – Median value ", fontsize=12)

sns.despine()
plt.tight_layout()
plt.savefig("Rooms vs. Median Home Value.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10, 7))


sns.histplot(
    data=df, x="MEDV", kde=True,
    color="#378ADD",
    edgecolor="white",
    linewidth=0.6,
    alpha=0.75,
)
plt.title("Distribution of Median Home Value", fontsize=15, pad=15, fontweight="bold")
sns.despine()
plt.savefig("Distribution of Median Home Value.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python


plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=df, x="LSTAT", y="MEDV",
    s=40, alpha=0.55, edgecolor="white", linewidth=0.5
)

sns.regplot(
    data=df, x="LSTAT", y="MEDV",
    scatter=False, lowess=True,
    line_kws={"color": "#D81B60", "lw": 2}
)
plt.title("LSTAT vs MEDV", loc="center", weight="bold")
plt.xlabel("LSTAT — Percentage of lower socio-economic status")
plt.ylabel("MEDV — Median home value ")

sns.despine()
plt.tight_layout()
plt.savefig("LSTAT vs MEDV.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.boxplot(x='CHAS', y='MEDV', data=df)

plt.title("House Price vs Charles River (CHAS)")
plt.xlabel("Near River (No = 0, Yes = 1)")
plt.ylabel("Median House Price (MEDV)")
plt.savefig("House Price vs Charles River (CHAS).png", dpi=300, bbox_inches='tight')

plt.show()
```
